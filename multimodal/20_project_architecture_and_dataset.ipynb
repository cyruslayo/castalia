## Step 1: Lay out the system components

The capstone should feel like one coherent system, not a bag of disconnected demos.


```python
architecture = pd.DataFrame([
    {"component": "vision router", "responsibility": "decide between screenshot QA, document path, and video path"},
    {"component": "document retriever", "responsibility": "retrieve relevant pages with text and visual evidence"},
    {"component": "audio processor", "responsibility": "transcribe or classify audio clips"},
    {"component": "video processor", "responsibility": "sample frames and summarize short clips"},
    {"component": "report builder", "responsibility": "merge answers with evidence for handoff"},
])
architecture
```

## Step 2: Define the dataset schema

A single schema keeps the capstone benchmark manageable even when the inputs differ by modality.


```python
example_record = {
    "example_id": "cap-001",
    "task_type": "video_summary",
    "inputs": {"video_path": "demo.mp4", "prompt": "Explain the failure in the clip."},
    "expected_answer": "The password reset flow fails after submit.",
    "expected_evidence": "5.0-8.0s",
    "difficulty": "medium",
}
print(json.dumps(example_record, indent=2))
```

## Step 3: Organize the artifacts

Capstones become reusable once the artifact layout is consistent.


```python
artifact_layout = pd.DataFrame([
    {"artifact": "dataset.jsonl", "purpose": "benchmark examples"},
    {"artifact": "retrieval_runs.csv", "purpose": "retrieval and grounding traces"},
    {"artifact": "eval_report.csv", "purpose": "aggregate metrics"},
    {"artifact": "handoff.md", "purpose": "human-readable summary"},
])
artifact_layout
```
